# Label Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["awscube", "cow", "dog", "cat", "black"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()                          # calling LabelEncoder
df["en_name"] = le.fit_transform(df["name"])
```

```python
df
```

```python
dataset = pd.read_csv("california_cities.csv")
dataset.head()
```

```python
la = LabelEncoder()
la.fit(dataset["city"])
```

```python
dataset["city"] = la.transform(dataset["city"])
```

```python
dataset
```

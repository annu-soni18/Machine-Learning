# How to find missing/null values/data in Dataset

```python
port numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print(np.__version__)
```

```python
dataset = pd.read_csv("california_cities.csv")
dataset.head(3)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()     # kitne mere null/missing values/data h wo bta rha h column-wise
```

```python
dataset.shape[0]            # Total no of rows inn dataset
```

```python
dataset.isnull().sum().sum()            # t0tal no of null values in entire dataset
```

```python
dataset.isnull().sum()/dataset.shape[0]*100     # column-wise percentage of  null values
```

```python
(dataset.isnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]))*100              #percentage of null values in entire dataset
```

```python
dataset.notnull().sum()      # Total no of filled data column-wise
```

```python
dataset.notnull().sum().sum()          # Total no of filled data in entire dataset
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

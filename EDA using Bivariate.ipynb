```python
de = pd.read_csv('train (1).csv')
```

```python
de.head()
```

### 3. Box Plot (Numerical - Categorical)

```python
sns.barplot(data=de, x='Pclass', y='Age',hue='Sex')

# here hue part is showing the relation between x and y to analyse data more closely
```

### 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data = de, x='Sex', y='Age', hue = 'Survived')          

# here we have analyse multiple columns together to see their relation, and analyse the data more closely
```

### 4. Distplot (Numerical - Categorical)

```python
sns.distplot(data = de, x='Age', y= 'Fare')
```

# 1) Calculate the multiplication and sum of two numbers

```python

def multiply_and_sum(a, b):
    multiplication = a * b
    summation = a + b
    print(f"Multiplication: {multiplication}")
    print(f"Sum: {summation}")

multiply_and_sum(5, 3)
```

# 2) Print the sum of the current number and the previous number

```python
def print_sum_current_previous(n):
    previous = 0
    for current in range(n):
        total = current + previous
        print(f"Current Number: {current}, Previous Number: {previous}, Sum: {total}")
        previous = current

print_sum_current_previous(10)
```

# 3) Remove first n characters from a string

```python
def remove_first_n_chars(string, n):
    result = string[n:]
    print(f"String after removing first {n} characters: {result}")

remove_first_n_chars("HelloWorld", 5)
```

# 4)  Check if the first and last numbers of a list are the same

```python
def check_first_last_same(numbers):
    if numbers[0] == numbers[-1]:
        print("True: First and last elements are the same")
    else:
        print("False: First and last elements are different")

check_first_last_same([10, 20, 30, 40, 10])
```

# 5) Display numbers divisible by 5

```python
def divisible_by_5(numbers):
    print("Numbers divisible by 5:")
    for num in numbers:
        if num % 5 == 0:
            print(num)

divisible_by_5([10, 23, 45, 66, 75, 89])
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

nltk.download('punkt')
nltk.download('stopwords')

text = input("Enter the passage: ")
tokens = word_tokenize(text)
stop_words = set(stopwords.words("english"))
filtered = [word for word in tokens if word.lower() not in stop_words]
print("Filtered Words:", filtered)
```

```python
from nltk.corpus import stopwords
import nltk
import re

nltk.download('stopwords')

text = input("Enter the passage: ")
tokens = re.findall(r'\b\w+\b', text)

stop_words = set(stopwords.words("english"))
filtered = [word for word in tokens if word.lower() not in stop_words]

print("Filtered Words:", filtered)
```

```python
from nltk.stem import LancasterStemmer
from nltk.corpus import stopwords
import nltk
import re


nltk.download('stopwords')


stemmer = LancasterStemmer()
stop_words = set(stopwords.words("english"))


sample = input("Enter the Statement: ")


tokens = re.findall(r'\b\w+\b', sample)
print("Tokens:", tokens)


for word in tokens:
    if word.lower() not in stop_words:
        print(word, "->", stemmer.stem(word))
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

sample = input("Enter the Statement: ")

token_data = word_tokenize(sample)
print("Tokens:", token_data)

pos_tags = pos_tag(token_data)

print("POS Tags:")
for word, tag in pos_tags:
    print(f"{word}: {tag}")

```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import LancasterStemmer
nltk.download("punkt")
nltk.download("punkt_tab")
Sample = input("Enter the Statement: ")
token_data = word_tokenize(Sample)
print("Tokens:", token_data)
port=LancasterStemmer()
for i in token_data:
    print(i + " " + port.stem(i))

```

```python
import nltk
import random
from nltk.corpus import movie_reviews
nltk.download('movie_reviews')
documents = [(list(movie_reviews.words(fileid)), category) 
             for category in movie_reviews.categories() 
             for fileid in movie_reviews.fileids(category)]

random.shuffle(documents)
print("\nSample Document (Words and Category):\n")
print(documents[1])
all_words = [w.lower() for w in movie_reviews.words()]
all_words = nltk.FreqDist(all_words)
print("\nMost Common Words:\n")
print(all_words.most_common(15))
print("\nFrequency of 'movie':", all_words["movie"])


```
